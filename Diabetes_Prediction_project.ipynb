Importing the dependencies

```python
import numpy as np
import pandas as pd #
from sklearn.preprocessing import StandardScaler #Standerdizes the data into a common range and the sklearn.preprocessing is a module that provides tools to prepare the data before feeding it into a maachine learing model/ cleaning and grooming
from sklearn.model_selection import train_test_split
from sklearn import svm #SVM- Support vector machine
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

PIMA Diabetes Datasets - info of patients that have diabetes, for females.

```python
# loading the diabetes dataset ro a pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes.csv') #command separated value
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# separating the data and labels
X= diabetes_dataset.drop(columns = 'Outcome',axis=1)
Y= diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standardisation

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standerdized_data = scaler.transform(X)
```

```python
print(standerdized_data)
```

```python
X = standerdized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

Training the model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the traning data: ', training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data: ',test_data_accuracy) #war fiiting?
```

Making a Predictive system

```python
input_data = (4,110,92,0,0,37.6,0.191,30)

#change the data into numpy array- a more easy and efficient processing on the numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the array as we are predicting for one instance 47:17mins
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardizing the input data
std_data = scaler.transform(input_data_reshaped)

prediction = classifier.predict(std_data)

if(prediction[0]=='Patient has no diabetes'):
  print('The object is a rock')
else:
  print('Patient has diabetes')
```
